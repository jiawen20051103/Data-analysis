```python
import pandas as pd
```

```python
user_level = pd.read_excel('user_level.xlsx')
```

```python
import os
```

```python
excel_list = []
for item in os.listdir('./'):
    if 'xlsx' in item and 'user_level' not in item:
        excel_list.append(item)
excel_list
```

```python
danmu = pd.DataFrame()
for item in excel_list:
    excel = pd.read_excel(item, converters={'uid': str, 'id': str})
    excel[''] = item
    danmu = pd.concat([danmu, excel], axis=0)
danmu
```

```python
from datetime import datetime
```

```python
danmu[''] = danmu['ctime'].map(datetime.fromtimestamp)
```

```python
danmu[''] = danmu[''].map(lambda x: x.year)
```

```python
danmu[''] = danmu[''].map(lambda x: x.month)
danmu[''] = danmu[''].map(datetime.isoweekday)
danmu[''] = danmu[''].map(lambda x: x.hour)
danmu
```
