# 1.数据清洗与处理

```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
bill = pd.read_excel('bill.xlsx')
bill
```

新增 月份、日期、星期、时段 

```python
bill['月份'] = bill['交易时间'].dt.month
bill['日期'] = bill['交易时间'].dt.date
bill['星期'] = bill['交易时间'].dt.weekday+1  #dt.weekday获取星期数0-6 +1转换成1-7
def cal(hour):
    if 6<=hour<8:
        return '早上'
    elif 8<=hour<11:
        return '上午'
    elif 11<=hour<14:
        return '中午'
    elif 14<=hour<19:
        return '下午'
    elif 19<=hour<24:
        return '晚上'
    else:
        return '凌晨'
bill['时段'] = bill['交易时间'].dt.hour.apply(cal)
bill
```

```python
# 把‘交易对方’中的‘慧生活798’提取到新的文件中
water = bill[bill['交易对方']=='慧生活798']  #筛选
water.to_excel('water.xlsx',index=False)  #保存
bill = bill[bill['交易对方']!='慧生活798']  #删除相关数据
bill = bill.reset_index(drop=True)  #重新生成索引号
bill.to_excel('bill.xlsx',index=False)
bill
```

# 2.基础收支统计

```python
# 将'/'变成'不计收支'
bill.loc[bill['收支类型']=='/','收支类型'] = '不计收支'
sum_of_bill = bill.groupby('收支类型')['金额(元)'].sum().reset_index()
result = sum_of_bill.iloc[[1,2]]  # 仅显示收入、支出两项
result
```

# 3.收支结构分析

## 3.1 支出结构分析

```python
bill_out = bill[bill['收支类型']=='支出']
# 统计排序，选取前十条数据
catagorials = bill_out.groupby('交易类型')['金额(元)'].sum().reset_index().sort_values('金额(元)',ascending=False)[:10]
# 绘制饼图
plt.pie(catagorials['金额(元)'],labels=catagorials['交易类型'],autopct='%1.1f%%')
plt.title('支出结构占比')
plt.show()
```

## 3.2 收入结构分析

```python
bill_in = bill[bill['收支类型']=='收入']
# 统计排序，选取前五条数据
catagorials = bill_in.groupby('交易类型')['金额(元)'].sum().reset_index().sort_values('金额(元)',ascending=False)[:5]
# 绘制饼图
plt.pie(catagorials['金额(元)'],labels=catagorials['交易类型'],autopct='%1.1f%%')
plt.title('收入结构占比')
plt.show()
```

# 4.消费时间分析

## 4.1 消费时段

```python
# 统计各时段的消费次数和总金额
period_sum = bill_out.groupby('时段').agg(消费次数=('金额(元)','count'),总金额=('金额(元)','sum')).sort_values('总金额',ascending=False).reset_index()
plt.pie(period_sum['总金额'],labels=period_sum['时段'],autopct='%1.1f%%')
plt.title('消费时段总金额占比')
plt.show()
```

消费时段总金额占比比较均匀，除了早上和凌晨，其他时段消费金额占比相近。那如果仅看消费次数会是怎样的分布呢？

```python
# 统计各时段的消费次数和总金额
period_count = bill_out.groupby('时段')['金额(元)'].agg(消费次数='count').reset_index()
plt.pie(period_count['消费次数'],labels=period_count['时段'],autopct='%1.1f%%')
plt.title('消费时段消费次数占比')
plt.show()
```

若仅看消费次数，还是下午占比最大，晚上和中午占比相近

## 4.2 月份

```python
month_sum =  bill_out.groupby('月份').agg(消费次数=('金额(元)','count'),总金额=('金额(元)','sum')).sort_values('总金额',ascending=False).reset_index()
# 绘制柱状图
plt.bar(month_sum['月份'],month_sum['总金额'])
# 设置图表标题和坐标轴标签
plt.title('消费月份总金额分布')
plt.xlabel('月份')
plt.ylabel('总金额')
plt.show()
```

八月份支出金额最多，一方面是要交学费，另一方面可能是开学买东西

## 4.3 星期

```python
week_sum =  bill_out.groupby('星期').agg(消费次数=('金额(元)','count'),总金额=('金额(元)','sum')).sort_values('总金额',ascending=False).reset_index()
# 绘制柱状图
plt.bar(week_sum['星期'],week_sum['总金额'])
# 设置图表标题和坐标轴标签
plt.title('消费星期总金额分布')
plt.xlabel('星期')
plt.ylabel('总金额')
plt.show()
```

整体呈现缓慢下行趋势，但周三格外高，我猜测是有一笔大的支出在周三进行。为验证此想法可以找出最大的十笔消费记录进行验证

```python
# 筛选出支出记录
expense_bill = bill[bill['收支类型'] == '支出']
# 找出最大的十笔支出记录
top_ten_expenses = expense_bill.nlargest(10, '金额(元)')
print(top_ten_expenses)
```

一笔5700元的支出在周三，说明推测正确

```python
week_count =  bill_out.groupby('星期').agg(消费次数=('金额(元)','count')).sort_values('消费次数',ascending=False).reset_index()
# 绘制柱状图
plt.bar(week_count['星期'],week_count['消费次数'])
# 设置图表标题和坐标轴标签
plt.title('消费星期总次数分布')
plt.xlabel('星期')
plt.ylabel('消费次数')
plt.show()
```

就消费次数而言，周五、周六、周日消费次数更高，可能与周末假期有关

## 4.4 关联事件

```python
thing_count =  bill_out.groupby('关联事件').agg(消费次数=('金额(元)','count')).sort_values('消费次数',ascending=False).reset_index()
# 绘制柱状图
plt.bar(thing_count['关联事件'],thing_count['消费次数'])
# 设置图表标题和坐标轴标签
plt.title('关联事件消费总次数分布')
plt.xlabel('关联事件')
plt.ylabel('消费次数')
plt.show()
```

还是在学校的消费次数多，可能是因为放假在家不用自己花钱。也有可能与关联事件所对应的天数有关。可以筛选出各个关联事件对应的天数来加以判断。

```python
bill.groupby('关联事件')[['日期']].nunique().plot(kind='bar')
```

由此可见，消费次数是与关联事件天数成正比的

# 5.可优化消费分析

## 5.1 零食饮料消费分析

```python
snack = bill[bill['交易类型']=='零食饮料']
snack
```

```python
# 零食饮料中支出类型的总金额
sum_snack = snack[snack['收支类型']=='支出']['金额(元)'].sum()
# 结果保留一位小数
round(sum_snack,1)
```

### 5.1.1 每月消费次数分布折线图

```python
import numpy as np
# 每月消费次数分布
daily_counts = snack.groupby('月份').size()
# 创建一个新的图形窗口
plt.figure(figsize=(12,4))
daily_counts.plot(kind='line', marker='o', color='#FF6B6B')
plt.title('每月零食饮料消费次数趋势')
# 绘制水平参考线
plt.axhline(y=np.mean(daily_counts), color='gray', linestyle='--', label='均值')
```

4-7月、9-10月、12月零食饮料消费超出平均值，尤其是5月和10月，更是高达22次！可能跟在校有关。

### 5.1.2 消费时段分布热力图

```python
import seaborn as sns
# 消费时段分布热力图
# 创建一个筛选结果副本，防止后续对snack的操作影响到原始的bill数据集，同时也避免了SettingWithCopyWarning警告。
snack = bill[bill['交易类型']=='零食饮料'].copy()
snack['小时'] = snack['交易时间'].dt.hour
hour_heatmap = snack.pivot_table(index='小时', values='金额(元)', aggfunc='count')
sns.heatmap(hour_heatmap, cmap='YlOrRd', annot=True, fmt='d')
plt.title('消费时段分布热力图')
# 添加 x 轴标签
plt.xlabel('消费次数')
plt.show()
```

16-19这个时段消费比较集中，可能是下午跟同学出去玩

## 5.2 支付方式优化

### 5.2.1 支付方式与金额分布箱线图

```python
# 支付方式与金额分布箱线图
sns.boxplot(x='支付方式', y='金额(元)', data=bill, hue = '支付方式', palette='Pastel1', legend = False)
plt.title('不同支付方式的单笔金额分布')
```

微信支付的大额消费要多于支付宝，两者在中低额度消费场景表现稳定且集中，也能表明平时用微信支付更多。

## 5.3 模拟减少20%深夜消费

```python
# 模拟减少20%深夜消费的节省效果
bill['小时'] = bill['交易时间'].dt.hour
night_spend = bill[bill['小时'].between(22,24)]['金额(元)'].sum()
simulations = [night_spend * 0.2 * np.random.normal(1, 0.1) for _ in range(1000)]
# f-string格式化输出年度预估节省金额的均值和标准差
print(f"年度预估节省：{np.mean(simulations):.2f}±{np.std(simulations):.2f}元")
```

看我是否做到了每天两千毫升饮水量，已知我的被子有700ml，所以我应每天至少饮水三次才能达到要求。
